2\. Convert the following function into a pure function with no global variables or side effects

````python
x = 5
def f(alist):
    for i in range(x):
         alist.append(i)
    return alist

alist = [1,2,3]
ans = f(alist)
print (ans)
print (alist) # alist has been changed!
````

````python
# Solution 2

import copy

def f(alist):
    new_list = copy.deepcopy(alist)
    for i in range(5):
         new_list.append(i)
    return new_list

alist = [1,2,3]
ans = f(alist)
print (ans)
print (alist) # alist has not been changed!
````

3\. Write a `decorator` hello that makes every wrapped function print “Hello!”, i.e. something like:

```python
@hello
def square(x):
    return x**2
```



````python
def hello(func): #el decorador recibe una función como argumento
    def wrapper(*args, **kwargs):
        print("Hello!")
        return func(*args, **kwargs)
    return wrapper


@hello # el símbolo @ es solo una forma corta para aplicar una función decoradora a otra función.
def square(x):
    return x**2

print(square(5))
````

4\. Write the factorial function so that it a) does and b) does not use recursion.

````python
# Solution 4a

def factorial_1(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial_1(n-1)
        
factorial_1(5)
````

````python
# Solution 4b

def factorial_2(n):
    i = 1 # inicia un "contador" en 1 para empezar a multiplicar
    for x in range(1,n+1): # si se pone range(n) incluye el cero
        i = i * x  # multiplica el "contador" por el número actual de la iteración
    return i     
        
factorial_2(5)
````

6\. Edit the class defintion to add an instance attribute of is_hungry = True to the Dog class. Then add a method called eat() which changes the value of is_hungry to False when called. Figure out the best way to feed each dog and then output “My dogs are hungry.” if all are hungry or “My dogs are not hungry.” if all are not hungry. The final output should look like this:

`I have 3 dogs. 
Tom is 6. 
Fletcher is 7. 
Larry is 9.

````python
# Parent class
class Dog:

    # Class attribute
    species = 'mammal'

    # Initializer / Instance attributes
    def __init__(self, name, age, is_hungry):
        self.name = name
        self.age = age
        self.is_hungry = True # 1

    # instance method
    def description(self):
        return "{} is {} years old".format(self.name, self.age)

    # instance method
    def speak(self, sound):
        return "{} says {}".format(self.name, sound)

    # instance method
    def eat(self):
        self.is_hungry = False # 2

# Child class (inherits from Dog class)
class RussellTerrier(Dog):
    def run(self, speed):
        return "{} runs {}".format(self.name, speed)

# Child class (inherits from Dog class)
class Bulldog(Dog):
    def run(self, speed):
        return "{} runs {}".format(self.name, speed)
````

````python
# esta celda es de prueba

toby = Dog("toby",15,False)

print(toby.name)
print(toby.age)
print(toby.is_hungry)

toby.eat()
print(toby.is_hungry)
````

````python
# Solution 6

tom = Dog("Tom", 6, False)
fletcher = Dog("Fletcher", 7, True)
larry = Dog("Larry", 9, True)

# notice that it does not matter if i set the is_hungry argument to false, they are all going to be hungry
# because the __init__ sets self.is_hungry = True regardless of the argument.

# So, now I want to feed arbitrarily one of the dogs
import random
dogs = [tom, fletcher, larry]
dogs_to_feed = random.sample(dogs, 2)

for dog in dogs_to_feed:
    dog.eat()

# check which dog is hungry
print('Is my dog Tom hungry?',  tom.is_hungry)
print('Is my dog Fletcher hungry?', fletcher.is_hungry)
print('Is my dog Larry hungry?', larry.is_hungry)
print(' ')

# check again
for dog in dogs:
    if dog not in dogs_to_feed:
        dog.eat()
        print('Is my dog', dog.name, 'still hungry?', dog.is_hungry)
````

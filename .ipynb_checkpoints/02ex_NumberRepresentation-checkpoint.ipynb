1\. Write a function that converts number representation, bin<->dec<->hex. (Clearly using the corresponding python built-in functions is not fair..)

```python
def dec_bin(x):

    bin_list = [] # lista vacia donde vamos a guardar los digitos (0s o 1s) del resultado
    num = int(x) # convertimos la entrada de la función en entero (porque no sé como hacerlo con punto decimal aun)

    while num > 0: # mientras que el número sea mayor que cero hace este loop. OJO: num va a ir cambiando
        residuo = num%2 # calculamos el residuo de la primera devisión entre 2
        bin_list.append(residuo) # lo agregamos a la lista
        num = num // 2 # actualizamos num para que ahora sea el cociente de la división (sin decimales)
        
    return bin_list[::-1] # invertimos la lista
```

```python
dec_bin(13) #1101
```

```python
def bin_dec(x):

    num = str(x).zfill(8)  # 8 dígitos, rellena con ceros a la izquierda
    bin_digits = [int(digit) for digit in num]
    bin_exponents = [2**7, 2**6, 2**5,  2**4,  2**3,  2**2, 2**1, 2**0]

    prueba = list(zip(bin_digits,bin_exponents))
    final = [i*j for i,j in prueba]

    return sum(final)
```

```python
bin_dec(1101)
```

```python
def dec_hex(x):
    hex_list = []
    num = int(x)

    if num<=9:
        return num
    elif num==10:
        return A
    elif num==11:
        return B
    elif num==12:
        return C
    elif num==13:
        return D
    elif num==14:
        return E
    elif num==15:
        return F
    else:
        while num > 0:
            residuo = num%16
            if residuo==10:
                hex_list.append('A')
            elif residuo==11:
                hex_list.append('B')
            elif residuo==12:
                hex_list.append('C')
            elif residuo==13:
                hex_list.append('D')
            elif residuo==14:
                hex_list.append('E')
            elif residuo==15:
                hex_list.append('F')
            else:
                hex_list.append(residuo)
            num = num // 16
        
    return hex_list[::-1]
```

2\. Write a function that converts a 32 bit word into a single precision floating point (i.e. interprets the various bits as sign, mantissa and exponent)

```python
def ttbits_interpretation(ttbits):
    
    # extraer el signo
    s = ttbits >> 31 # se mueve 31 a la derecha para descartar todos los bits excepto el primero

    # extraer el exponente (8 bits)
    e = ttbits >> 1 # se mueve uno a la izquiera para descartar el bit del signo
    e = ttbits >> 23 # se nueve 23 a la derecha para descartar los bits de la mantissa

    # extraer la mantissa (23)
    m = ttbits << 9 # se mueve a la izquierda para descartar los primeros 9 bits
    
    return ((-1)**s) * (0+m) * (2**(e-127))

# PENDIENTE
```

```python
ttbits_interpretation(00000011111000000000000000000000)
```

7\. Consider the integral of the semicircle of radius 1:

$$

with $h=2/N$ the width of each of the $N$ slices the domain is divided into and where
$y_k$ is the value of the function at the $k-$th slice.
